## Student Practice 1: Write a Function
Write a function named `calculate_average()` that takes a list of numbers and returns the average.

## Student Practice 2: Use the Function
Use your function to calculate the average of `[10, 20, 30, 40]`.

```python
# Completed Answer 1 and 2: Function Practice
def calculate_average(numbers):
    return sum(numbers) / len(numbers)

values = [10, 20, 30, 40]
result = calculate_average(values)
print('Average:', result)
```
