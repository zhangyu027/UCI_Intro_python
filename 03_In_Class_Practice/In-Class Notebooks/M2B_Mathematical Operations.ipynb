# M2_Mathematical Operations

## We just saw a list of operators in the slides. Now, look at the examples below,
and please take time to edit them and write your own equations.


```python
# Let's do some math with integers and floats

# addition and subtraction use + and -:
# with integers, we get integer answers:
print('\nAddition and Subtraction with Integers:')

answer = 10 - 2
print(answer)
print(type(answer))

answer = 10 + 2
print(answer)
print(type(answer))

print('\nAddition and Subtraction with Floating Numbers:')
# with floating numbers, we get floating number answers:
answer = 10.0 - 2.0 + 5.0
print(answer)
print(type(answer))

print('\nAddition and Subtraction with Floating Numbers and Integers:')
# When you mix floating numbers and integers, the answer will be a floating
# number.
answer = 10 - 2 + 5.0
print(answer)
print(type(answer))

print('\nMultiplication and Division with Intengers:')
# when multiplying with integers, your result will be an integer.
# when dividing with integers, your result will be a floating point number,
# even if the answer is a whole number.
# multiplication uses the '*' character:
answer = 3 * 5
print(answer)
print(type(answer))

# division uses the '/' character:
answer = 6 / 3
print(answer)
print(type(answer))

print('\nMultiplication and Division with Floats:')
answer = 5.0 * 5.765
print(answer)
print(type(answer))

answer = 5.0 / 2.123
print(answer)
print(type(answer))

print('\nMultiplication and Division with a mixture of Integers and Floats '
      'will result in a float')

answer = 7 * 5.0 / 2.123 * 6
print(answer)
print(type(answer))

print('\nThe modulo operator %')

answer = 10 % 7
print(answer)
print(type(answer))

print('\nInteger division // (or division without the remainder)')

answer = 10.0 // 7.0
print(answer)
print(type(answer))

print('\nRaising a number to a power')

answer = 3**2
print(answer)
print(type(answer))

print('\nFinding the square root of a number')

answer = 9**(1/2)
print(answer)
print(type(answer))
```

## Comparison Operators
Comparison operators are what we use to generate booleans.  They also help determine how our script or program flows - meaning if a certain value is positive, our script may do one thing, if it is negative, it may do something else.

For example, you may want to read in transaction data, and write out the transactions that are above a certain amount, you need comparison operators to determine if the transaction amount is greater than or less than your threshold value.

Let's see some examples below:

```python
Mynum = 5
```

```python
# We can use == to check if two values are equal:
print('Sophie' == 'Sophie')

# We can check if two values are not equal, using !=
print('Sophie' != 'Beth')

# We can check if a value is greater than, or less than, using > or <
print(5 > 7)
print(9 < 10)

# We can also use 'greater than or equal to, >=, or 'less than or equal to' <=
print(2 <= 3)


# Now add some of your own comparisons below, and print the boolean results

```

## And / Or
We can also check if two thing are true, and we check if one of two (or more things) are true. We do this by using the 'and' and 'or' keywords.

```python
myBoolean = (2 > 1) and (7 > 6)
print(myBoolean)


myBoolean = (-5 < 1) or (7 < 2)
print(myBoolean)


# In the space below, combine the comparison operators we learned, with and/or
# to create some more complex logical statements.
```

## Student Practice 1: Math Operations
A student earns 85, 92, and 78 on three assignments. Calculate the total and average.

## Student Practice 2: Percent Change
A price increases from 40 to 50. Calculate the percent increase.

```python
# Completed Answer 1: Total and Average
assignment_1 = 85
assignment_2 = 92
assignment_3 = 78

total = assignment_1 + assignment_2 + assignment_3
average = total / 3
print('Total:', total)
print('Average:', average)

# Completed Answer 2: Percent Increase
old_price = 40
new_price = 50
percent_increase = ((new_price - old_price) / old_price) * 100
print('Percent increase:', percent_increase, '%')
```
