```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('UCI_CDE_Capstone_Sales_Dataset.csv')
print(df.head())
print(df.describe())
print(df.isna().sum())
sales=df.groupby('Region')['Sales'].sum().sort_values()
sales.plot(kind='barh')
plt.title('Sales by Region')
plt.show()
print(df.groupby('Category')[['Sales','Profit']].sum())
```
