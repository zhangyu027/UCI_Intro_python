# M2 - Tuples

## Tuples
Tuple are like lists except they are immutable meaning you can not change them once you have created them. Creating a tuple is like creating a list, except we use round bracket () instead of square brackets []. One advantage tuples have is 'packing' and 'unpacking'. We will see some examples of this below.

```python
# First let's make a tuple!
myTuple = (5, 'Apples')
print(myTuple)
```

```python
# Now, let's try and update it - remember this will return an error, because we can't do this.
myTuple[0] = 1
```

## Tuple Unpacking

Tuple unpacking is a convenient way to assign elements of a tuple to multiple variables at once. The number of variables on the left-hand side of the assignment must match the number of elements in the tuple. This is particularly useful when a function returns a tuple, or when you want to swap variable values easily.

```python
# Example of tuple unpacking
coordinates = (10, 20)
x, y = coordinates
print(f"x: {x}, y: {y}")

# Another example: swapping values
a = 5
b = 10
print(f"Before swap: a={a}, b={b}")
a, b = b, a # Unpacking with a temporary tuple (b, a)
print(f"After swap: a={a}, b={b}")

# Unpacking from a function's return value
def get_user_info():
    return ("Alice", 30, "New York")

name, age, city = get_user_info()
print(f"User Info: Name={name}, Age={age}, City={city}")
```

## Dictionary
Dictionaries allow us to store data in key value pairs.  This means that we can label the data elements in the dictionary. For example, we could create a dictionary that contains user informations, like name, age, and state.

```python
# We can define a dictionary like seen below.  Note the use of curly brackets,
# {}, and the ':' used to separate the key and value.

# This dictionary has 3 pieces of data, the keys are 'name', 'age' and 'state.
# The values are show below.
newDict = {'name': 'Jill Smith', 'age': 35, 'state': 'OR'}

# We can access the data using the keys
userName = newDict['name']
print(userName)

print(newDict)

# We can remove items from the dictionary using 'del', just like we did with lists
del newDict['age']
print(newDict)

# We can add items to a dictionary in the following way:
newDict['height'] = 5.1
print(newDict)

nameList =['will', 'jill']
newDict['name'] = nameList[1]
print(newDict)

tempDict = {}
tempDict[(11.0, 45.0)] = 75.4
print(tempDict)
```

```python
# Also, note that keys can be strings, tuples, or numbers
myDict = {}
myDict[(1, 2)] = 'apple'
print(myDict)
```

```python
# Speaking of iterator objects, we can create one using range.  Range(n) will
# generate numbers from 0 to n-1.  For example, range(5) returns a range object
# which we can use to iterate from 0 to 4.
# The idea of an iterator may be confusing, but just think if it as a list,
# except only one value in the list is in memeory at a time.  So instead of
# creating the list [0, 1, 2, 3, 4], you have created a range object that
# "counts" from 0 to 4, one number at a time.
range(5) # will generate numbers from 0 to 4
```

```python
# Can also 'skip' a loop in the for loop, but still continue the loop, using
# the 'continue' keyword.
for val in range(3):
    if val == 1:
        continue
    newVal = val**2
    print(newVal)
```

```python
# You can also exit a loop using the break keyword:
for word in ['apple', 'oregon', 'washington']:
    print(word)
    if word == 'oregon':
        break
```

# M2_If/Else
If/Else blocks are our first look at a topic called 'flow' control! Flow control refers to methods that allow our script/program to do one thing or another depending on the current state of the program. Note the colons at the end of each 'if', 'else', and 'elif' line and the indentation each if/elif/else statement - this is mandatory!


```python
# First Example:
myWord = 'basil'

if len(myWord) <= 3:
    print('short word')
else:
    print('long word')
```

```python
# For example, consider a script that determines shipping cost based on the
# weight of a package:
packageWeight = 8

if packageWeight < 10:
    cost = 5
elif packageWeight >= 10 and packageWeight < 15:
    cost = 7
else: # package must have a weight >= 15 if the above if/else blocks havennot been triggered.
    cost = 15

print('The shipping cost is ', cost)
```

```python
# Note that we can have as many 'elif' portions of an if/else statment as needed:
packageWeight = 40
if packageWeight < 10:
    cost = 5
elif packageWeight >= 10 and packageWeight < 15:
    cost = 7
elif packageWeight >= 15 and packageWeight < 20:
    cost = 10
elif packageWeight >= 20 and packageWeight < 35:
    cost = 12
else: # package must have a weight >= 35 if the above if/else blocks haven not been triggered.
    cost = 18

print('The shipping cost is ', cost)
```
