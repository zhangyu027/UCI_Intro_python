## Instructions

Complete the function below. The function takes two arguments:

1. `namesDF`: a pandas DataFrame containing a `Name` column  
2. `name`: the name to filter for

The function should return a new DataFrame that only contains rows where the `Name` column equals the `name` argument.

```python
def returnDataForOneName(namesDF, name):
    """
    Return rows from namesDF where the 'Name' column is equal to name.
    """
    return namesDF[namesDF['Name'] == name]
```

```python
# Import pandas for data analysis.
import pandas as pd
```
