```python
# Import the required libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## Part 1: Load the Dataset

We will use the built-in Seaborn `tips` dataset.

```python
# Load the tips dataset using seaborn.
# The try/except makes the notebook work even when the internet is unavailable.
try:
    tips = sns.load_dataset('tips')
except Exception:
    tips = pd.DataFrame({
        'total_bill': [16.99, 10.34, 21.01, 23.68, 24.59, 25.29, 8.77, 26.88, 15.04, 14.78,
                       10.27, 35.26, 15.42, 18.43, 14.83, 21.58, 10.33, 16.29, 16.97, 20.65,
                       17.92, 20.29, 15.77, 39.42, 19.82, 17.81, 13.37, 12.69, 21.70, 19.65],
        'tip': [1.01, 1.66, 3.50, 3.31, 3.61, 4.71, 2.00, 3.12, 1.96, 3.23,
                1.71, 5.00, 1.57, 3.00, 3.02, 3.92, 1.67, 3.71, 3.50, 3.35,
                4.08, 2.75, 2.23, 7.58, 3.18, 2.34, 2.00, 2.00, 4.30, 3.00],
        'sex': ['Female','Male','Male','Male','Female','Male','Male','Male','Male','Male',
                'Male','Female','Male','Male','Female','Male','Female','Male','Female','Male',
                'Male','Female','Female','Male','Male','Male','Male','Male','Male','Female'],
        'smoker': ['No','No','No','No','No','No','No','No','No','No',
                   'No','No','No','No','No','No','No','No','No','No',
                   'No','No','No','No','No','No','No','No','No','No'],
        'day': ['Sun','Sun','Sun','Sun','Sun','Sun','Sun','Sun','Sun','Sun',
                'Sun','Sun','Sun','Sun','Sun','Sun','Sun','Sun','Sun','Sat',
                'Sat','Sat','Sat','Sat','Sat','Sat','Sat','Sat','Sat','Sat'],
        'time': ['Dinner'] * 30,
        'size': [2,3,3,2,4,4,2,4,2,2,2,4,2,4,2,2,3,3,3,3,2,2,2,4,2,4,2,2,2,2]
    })

# Display the first 5 rows.
tips.head()
```

```python
# Show basic information about the dataset.
print('Dataset information:')
tips.info()

print('\nSummary statistics:')
print(tips.describe())

print('\nCount by day:')
print(tips['day'].value_counts())
```

## Part 2: Matplotlib Basics

```python
# Create a histogram of total_bill using matplotlib.
plt.figure(figsize=(8, 5))
plt.hist(tips['total_bill'], bins=20)
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a scatter plot of total_bill vs tip using matplotlib.
plt.figure(figsize=(8, 5))
plt.scatter(tips['total_bill'], tips['tip'])
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

## Part 3: Seaborn Visualizations

```python
# Create a seaborn scatterplot.
plt.figure(figsize=(8, 5))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='smoker')
plt.title('Total Bill vs Tip by Smoker Status')
plt.show()
```

```python
# Create a seaborn boxplot.
plt.figure(figsize=(8, 5))
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title('Total Bill Distribution by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()

# Answer: Saturday and Sunday appear to have higher median total bills than Thursday and Friday.
# In this dataset, Sunday is often slightly higher depending on visual reading of the boxplot.
```

```python
# Create a seaborn barplot showing average tip by day.
plt.figure(figsize=(8, 5))
sns.barplot(data=tips, x='day', y='tip', estimator=np.mean, hue='day', legend=False)
plt.title('Average Tip by Day')
plt.xlabel('Day')
plt.ylabel('Average Tip')
plt.show()
```

```python
# Create a seaborn countplot showing how many observations are in each day.
plt.figure(figsize=(8, 5))
sns.countplot(data=tips, x='day', hue='day', legend=False)
plt.title('Number of Observations by Day')
plt.xlabel('Day')
plt.ylabel('Count')
plt.show()
```

## Part 4: Grouped Analysis and Visualization

```python
# Create a new dataframe with average total_bill and average tip by day.
day_summary = tips.groupby('day', observed=False)[['total_bill', 'tip']].mean().reset_index()
day_summary
```

```python
# Plot average tip by day using matplotlib.
plt.figure(figsize=(8, 5))
plt.bar(day_summary['day'].astype(str), day_summary['tip'])
plt.title('Average Tip by Day')
plt.xlabel('Day')
plt.ylabel('Average Tip')
plt.show()
```

## Part 5: Interpretation Questions

```python
# Q1: What relationship do you see between total_bill and tip?
# A1: There is a positive relationship. As the total bill increases, the tip usually increases too.

# Q2: Which chart is best for comparing distributions across days?
# A2: A boxplot is best because it shows the median, spread, and possible outliers for each day.

# Q3: Why might seaborn be easier than matplotlib for statistical plots?
# A3: Seaborn works directly with pandas DataFrames and can quickly create grouped statistical plots using arguments like x, y, and hue.
```
