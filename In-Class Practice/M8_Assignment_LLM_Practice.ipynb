## Part 1: Prompt Engineering Practice

```python
# COMPLETED ANSWER VERSION

# A weak prompt is vague and does not give much guidance.
weak_prompt = """
Explain linear regression.
"""

# A stronger prompt gives audience, task, format, and expected details.
strong_prompt = """
Explain linear regression to undergraduate students.
Include:
1. A simple one-paragraph definition
2. One real-world example
3. One limitation of linear regression
Use clear language and avoid advanced math.
"""

print("Weak Prompt:")
print(weak_prompt)

print("\nStrong Prompt:")
print(strong_prompt)
```

## Part 2: Simple Tokenization

Real LLM tokenization is more complex, but this simple example shows the basic idea.

```python
text = "Linear regression predicts a numeric value."

# Convert text to lowercase.
lowercase_text = text.lower()

# Split it into tokens using .split().
tokens = lowercase_text.split()

# Print the tokens and the number of tokens.
print("Original text:", text)
print("Lowercase text:", lowercase_text)
print("Tokens:", tokens)
print("Number of tokens:", len(tokens))
```

## Part 3: Zero-Shot vs Few-Shot Prompting

```python
# Create a zero-shot prompt that asks the model to classify this review.
review = "The chart was easy to understand, but the explanation was too short."

zero_shot_prompt = f"""
Classify the sentiment of the following student review as Positive, Negative, or Mixed.

Review: {review}

Sentiment:
"""

print(zero_shot_prompt)
```

```python
# Create a few-shot prompt with two examples before asking the model to classify the review.
few_shot_prompt = f"""
Classify each student review as Positive, Negative, or Mixed.

Example 1:
Review: The lecture was clear and the examples helped me understand.
Sentiment: Positive

Example 2:
Review: The homework was confusing and the instructions were missing details.
Sentiment: Negative

Now classify this review:
Review: {review}
Sentiment:
"""

print(few_shot_prompt)
```

## Part 4: Build a Tiny Retrieval System

Retrieval-Augmented Generation, or RAG, means retrieving relevant information first and then using that information to answer a question.

```python
documents = [
    "Linear regression predicts a continuous numeric outcome.",
    "Logistic regression predicts a category, often yes/no or 0/1.",
    "Decision trees split data using feature-based rules.",
    "A histogram shows the distribution of one numeric variable.",
    "A boxplot compares distributions across groups."
]

question = "Which model should I use for a yes or no prediction?"

# Use TfidfVectorizer and cosine_similarity from sklearn to retrieve the most relevant document.
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer = TfidfVectorizer()
document_vectors = vectorizer.fit_transform(documents)
question_vector = vectorizer.transform([question])

similarity_scores = cosine_similarity(question_vector, document_vectors).flatten()
best_index = similarity_scores.argmax()
retrieved_context = documents[best_index]

print("Question:", question)
print("Similarity scores:", similarity_scores.round(3))
print("Most relevant document:", retrieved_context)
```

```python
# Create a final prompt that includes retrieved context, the student question,
# and an instruction to answer using only the context.
final_prompt = f"""
Use only the context below to answer the student's question.

Context:
{retrieved_context}

Student question:
{question}

Answer in 2-3 clear sentences.
"""

print(final_prompt)
```

## Part 5: LLM Answer Evaluation

```python
llm_answer = "Use linear regression for yes or no prediction because it predicts categories."

# Evaluate the answer.
print("LLM Answer:", llm_answer)

is_accurate = False
explanation = "The answer is not accurate. Linear regression predicts continuous numeric values. Logistic regression is usually used for yes/no or 0/1 category prediction."

print("Is the answer accurate?", is_accurate)
print("Explanation:", explanation)
```

## Challenge Practice

```python
# Add two more documents to the documents list.
extended_documents = documents + [
    "A scatter plot shows the relationship between two numeric variables.",
    "Model accuracy measures how many classification predictions are correct."
]

# Ask a new question and retrieve the most relevant document.
new_question = "Which visualization should I use to compare two numeric variables?"

vectorizer2 = TfidfVectorizer()
extended_vectors = vectorizer2.fit_transform(extended_documents)
new_question_vector = vectorizer2.transform([new_question])

new_similarity_scores = cosine_similarity(new_question_vector, extended_vectors).flatten()
new_best_index = new_similarity_scores.argmax()
new_retrieved_context = extended_documents[new_best_index]

print("New question:", new_question)
print("Most relevant document:", new_retrieved_context)
print("Answer: Use a scatter plot because it shows the relationship between two numeric variables.")
```
