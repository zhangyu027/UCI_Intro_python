## Embedded Prompt: Dataset Orientation

Students may paste the following prompt into ChatGPT after they inspect the data manually.

**Prompt:**

> I am a student data analyst working with a retail dataset. The columns are: OrderID, OrderDate, Region, Category, CustomerSegment, Sales, Discount, Profit, and Quantity. The business problem is that sales are increasing but profitability is inconsistent. What are five useful first analyses I should perform? Do not invent results. Only suggest analyses and verification checks.

## Expected AI Response Themes

A strong AI response should suggest analyses such as:

- total sales and profit by region
- sales and profit by category
- monthly sales and profit trends
- discount impact on profit margin
- customer segment profitability
- missing value and duplicate checks
- outlier review

Instructor note: If AI invents numbers, use that as an immediate teachable moment.
